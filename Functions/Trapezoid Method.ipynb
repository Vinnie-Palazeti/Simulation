```python
import numpy as np
```

```python
def trapeZOID(f,a,b,N=50):
    x = np.linspace(a,b,N+1) 
    y = f(x)
    y_right = y[1:] 
    y_left = y[:-1]
    h = (b - a)/N
    area = (h/2) * np.sum(y_right + y_left)
    return area

trapeZOID(np.sin,0,np.pi/2,1000)

def foo(x):
  return 1 / (x*(np.log(10)- np.log(1)))

print("The probaility X < 5 via the trapezoid rule is",trapeZOID(foo,1,5,100))
```

```python
trapezoid_measures = []
bvalues = []

for i in range(3,26):
  def foo(x):
    return 1 / (x*(np.log(i)- np.log(1)))

  trapezoid_measures.append(trapeZOID(foo,1,(i/2)))
  bvalues.append(i)

print(trapeZOID(foo,1,3,N=10))

import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot()
ax.plot(bvalues,trapezoid_measures)
```
