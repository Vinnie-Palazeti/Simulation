```python
import numpy as np
import sys, os

class HiddenPrints:
    def __enter__(self):
        self._original_stdout = sys.stdout
        sys.stdout = open(os.devnull, 'w')

    def __exit__(self, exc_type, exc_val, exc_tb):
        sys.stdout.close()
        sys.stdout = self._original_stdout
```

```python

def Baccarat():
  
  player_wins = []
  banker_wins = []
  two_draw = []
  tie = []
  # list 1 - 13
  x = list(range(1,14))
  # repeat list to fill deck 
  deck = x * 4
  
  
  # intial roll
  playerc1 = int(np.random.choice(deck,1))
  deck.remove(playerc1)
  playerc2 = int(np.random.choice(deck,1))
  deck.remove(playerc2)
  

  # game condition
  playerhand = np.where(playerc1 in {10,11,12,13}, 0,playerc1) + np.where(playerc2 in {10,11,12,13}, 0,playerc2)

  # intial roll
  bankerc1 = int(np.random.choice(deck,1))
  deck.remove(bankerc1)
  bankerc2 = int(np.random.choice(deck,1))
  deck.remove(bankerc2)
  # game condition
  bankerhand = np.where(bankerc1 in {10,11,12,13}, 0,bankerc1) + np.where(bankerc2 in {10,11,12,13}, 0,bankerc2)



  # actual points
  if bankerhand > 9:
    bankerhand = bankerhand % 10
  else:
    pass

  if playerhand > 9:
    playerhand = playerhand % 10
  else:
    pass

  print("Banker hand is", bankerhand)
  print("Player hand is", playerhand)


  if playerhand in {8,9} and bankerhand in {8,9}:
    print("Both have an 8 or 9")
    print("The Banker was dealt", bankerc1,bankerc2,"with hand value", bankerhand)
    print("The Player was dealt", playerc1,playerc2,"with hand value", playerhand)
    two_draw.append(1)
  elif playerhand in {8,9}:
    print("Player wins!")
    print("The Player was dealt", playerc1,playerc2,"with hand value", playerhand)
    print("The Banker was dealt", bankerc1,bankerc2,"with hand value", bankerhand)
    two_draw.append(1)
  elif bankerhand in {8,9}:
    print("Banker wins!")
    print("The Banker was dealt", bankerc1,bankerc2,"with hand value", bankerhand)
    print("The Player was dealt", playerc1,playerc2,"with hand value", playerhand)
    two_draw.append(1)
  else:
    print("No winners in initial draw")
    print("The Banker was dealt", bankerc1,bankerc2,"with hand value", bankerhand)
    print("The Player was dealt", playerc1,playerc2,"with hand value", playerhand)
    print("Move to drawing rules")
    
    if playerhand in range(6):
      playerc3 = int(np.random.choice(deck,1))
      deck.remove(playerc3)
      playerhand = playerhand + np.where(playerc3 in {10,11,12,13}, 0,playerc3)
      print("Player drew", playerc3)
      #print("Player's new hand is", playerhand)

      if bankerhand in range(3):
        bankerc3 = int(np.random.choice(deck,1))
        deck.remove(bankerc3)
        bankerhand = bankerhand + np.where(bankerc3 in {10,11,12,13}, 0,bankerc3)
        print("Banker drew", bankerc3)
        #print("Banker's new hand is", bankerhand)

      elif bankerhand == 3 and playerc3 != 8:
        bankerc3 = int(np.random.choice(deck,1))
        deck.remove(bankerc3)
        bankerhand = bankerhand + np.where(bankerc3 in {10,11,12,13}, 0,bankerc3)
        print("Banker drew", bankerc3)
        #print("Banker's new hand is", bankerhand)

      elif bankerhand == 4 and playerc3 in range(2,8):
        bankerc3 = int(np.random.choice(deck,1))
        deck.remove(bankerc3)
        bankerhand = bankerhand + np.where(bankerc3 in {10,11,12,13}, 0,bankerc3)
        print("Banker drew", bankerc3)
        #print("Banker's new hand is", bankerhand)

      elif bankerhand == 5 and playerc3 in range(4,8):
        bankerc3 = int(np.random.choice(deck,1))
        deck.remove(bankerc3)
        bankerhand = bankerhand + np.where(bankerc3 in {10,11,12,13}, 0,bankerc3)
        print("Banker drew", bankerc3)
        #print("Banker's new hand is", bankerhand)

      elif bankerhand == 6 and playerc3 in range(6,8): 
        bankerc3 = int(np.random.choice(deck,1))
        deck.remove(bankerc3)
        bankerhand = bankerhand + np.where(bankerc3 in {10,11,12,13}, 0,bankerc3)
        print("Banker drew", bankerc3)
        #print("Banker's new hand is", bankerhand)

      else:
        print("No complex rules were met, so the Banker doesn't draw")
        pass

    else:
      print("Player's hand was",playerhand,", so they stand")
      if bankerhand in range(6):
        print("Banker's hand was",bankerhand,", so they draw")
        bankerc3 = int(np.random.choice(deck,1))
        deck.remove(bankerc3)
        bankerhand = bankerhand + np.where(bankerc3 in {10,11,12,13}, 0,bankerc3)
        print("Banker drew", bankerc3)
        #print("Banker's new hand is", bankerhand)
      else:
        print("Banker's hand was", bankerhand,", so they stand")
        pass

  if bankerhand > 9:
    bankerhand = bankerhand % 10
  else:
    pass

  if playerhand > 9:
    playerhand = playerhand % 10
  else:
    pass

  if bankerhand - playerhand == 0:
    print("Banker's final hand is", bankerhand)
    print("The Player's final hand is", playerhand)
    print("TIE")
    tie.append(1)

  elif bankerhand > playerhand:
    print("Banker's final hand is", bankerhand)
    print("The Player's final hand is,", playerhand)
    print("The Banker wins!")
    banker_wins.append(1)
  

  else:
    print("The Player's final hand is,", playerhand)
    print("Banker's final hand is", bankerhand)
    print("The Player wins!")
    player_wins.append(1)
  
  return banker_wins,player_wins, tie, two_draw

Baccarat() 
```

```python
def sim(N):
  ties = []
  playerwins = []
  bankerwins = []
  twodraws = []
  for i in range(N):
    bankerwin,playerwin,tie,twodraw = Baccarat()
    ties.append(sum(tie))
    playerwins.append(sum(playerwin))
    bankerwins.append(sum(bankerwin))
    twodraws.append(sum(twodraw))
  
  return sum(ties)/N, sum(playerwins)/N, sum(bankerwins)/N, sum(twodraws)/N

with HiddenPrints():
  prob = sim(100000)
```
