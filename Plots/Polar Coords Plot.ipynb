## Polar Coordinate Plot

(x(t), y(t)) has polar coordinates (sqrt(t), 2*pi*t)

Plot (x(t), y(t)) for t in [0,10]

```python
import numpy as np
import matplotlib.pyplot as plt

#X = np.linspace(-3,3,256,endpoint=True)
X = np.linspace(0, 10, 1000, endpoint=True)
x = np.sqrt(X)*(np.cos(2*np.pi*X))
y = np.sqrt(X)*(np.sin(2*np.pi*X))

fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot()
#projection='polar'
ax.plot(x,y, color = "red", lw = 3)
```
